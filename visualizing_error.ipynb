Visualizing error values for reservoir project

```python
#combining error csvs into one from ML_forecasting_error_reservoir  
import os
import pandas as pd

# Directory where the CSV files are located
csv_directory = 'Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH'

# List of CSV files in the directory
csv_files = [f for f in os.listdir(csv_directory) if f.startswith('Run2F')]

# Initialize an empty DataFrame
combined_df = pd.DataFrame()

# Load and concatenate each CSV file
for csv_file in csv_files:
    file_path = os.path.join(csv_directory, csv_file)
    df = pd.read_csv(file_path)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/combined_error_results_2F.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the combined error results CSV
error_results_df = pd.read_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/01_combined_error_results.csv')

# Select the MAE columns and the k values
mae_columns = ["1 MAE", "2 MAE", "3 MAE", "4 MAE", "5 MAE"]
k_values = error_results_df['k'].unique()

# Define custom colors for the bars
bar_colors = ["lightpink", "skyblue", "lightcoral", "lightgreen", "violet"]

# Calculate the width of each bar and the bar positions
bar_width = 0.15
bar_positions = np.arange(len(k_values))

# Create a bar plot for MAE vs. k
plt.figure(figsize=(18, 6))

for i, mae_column in enumerate(mae_columns):
    # Calculate the x positions for the grouped bars
    x_positions = bar_positions + i * bar_width
    
    # Plot the bars for each MAE column with custom colors
    plt.bar(x_positions, error_results_df.groupby('k')[mae_column].mean(), width=bar_width, label=mae_column, color=bar_colors[i])

# Set x-axis tick positions and labels
plt.xticks(bar_positions + (bar_width * (len(mae_columns) - 1) / 2), k_values)

# Set plot title and labels
plt.title('MAE across differnt K values and Forecast Horizons')
plt.xlabel('K Value')
plt.ylabel('MAE')

# Put the legend outside the plot
plt.legend(title='MAE', loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.savefig('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/month_.png')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the combined error results CSV
error_results_df = pd.read_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/015_combined_error_results.csv')

# Get the MAE columns and unique k values
mae_columns = ["1 MAE", "2 MAE", "3 MAE", "4 MAE", "5 MAE", "6 MAE"]
k_values = error_results_df['k'].unique()

# Calculate the width of each bar and the bar positions
bar_width = 0.07
bar_positions = np.arange(len(mae_columns))

# Define custom colors for the bars based on k values
bar_colors = plt.cm.viridis(np.linspace(0, 1, len(k_values)))

# Create a bar plot for MAE vs. forecast horizon
plt.figure(figsize=(16, 6))

for i, k in enumerate(k_values):
    # Calculate the x positions for the grouped bars
    x_positions = bar_positions + i * bar_width
    
    # Plot the bars for each k value with custom colors
    mae_values = [error_results_df[error_results_df['k'] == k][col].mean() for col in mae_columns]
    plt.bar(x_positions, mae_values, width=bar_width, label=f'K={k}', color=bar_colors[i])

# Set x-axis tick positions and labels
plt.xticks(bar_positions + (bar_width * (len(k_values) - 1)/2 ), mae_columns)


# Set plot title and labels
plt.title('MAE across different Forecast Horizons ')
plt.xlabel('Forecast Horizon (months)')
plt.ylabel('MAE (m)')

# Put the legend outside the plot
plt.legend(title='K Value', loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.savefig('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/MAE_1_140_skip.png')
plt.show()
```

```python
#linegraph of MAE across forecast horizon showing different k values
import pandas as pd
import matplotlib.pyplot as plt

# Load the combined error results CSV
error_results_df = pd.read_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/01_combined_error_results.csv')

# Get the MAE columns, unique k values, and the forecast horizons
mae_columns = ["1 MAE", "2 MAE", "3 MAE", "4 MAE", "5 MAE"]
k_values = error_results_df['k'].unique()
forecast_horizons = range(1, len(mae_columns) + 1)

# Define custom colors for the lines based on k values
line_colors = plt.cm.viridis(np.linspace(0, 1, len(k_values)))

# Create a line graph for MAE vs. forecast horizon
plt.figure(figsize=(10, 6))

for i, k in enumerate(k_values):
    mae_values = [error_results_df[error_results_df['k'] == k][col].mean() for col in mae_columns]
    plt.plot(forecast_horizons, mae_values, label=f'k={k}', marker='o', color=line_colors[i])

# Set plot title and labels
plt.title('MAE vs. Forecast Horizon')
plt.xlabel('Forecast Horizon')
plt.ylabel('MAE')

# Add a legend outside the plot
legend = plt.legend(title='k', loc='center left', bbox_to_anchor=(1, 0.5))
legend.set_title('k')

# Set x-axis ticks to be solid numbers [1, 2, 3, 4, 5]
plt.xticks(forecast_horizons)

# Show the plot
plt.grid(True)
plt.show()
plt.savefig('Z:/Research_Projects/Reservoir_Project/ML_outputs/tables_outputs_csv/error_results_per_FH/month_line.png')


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
df = pd.read_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/all_error.csv')

# Define the error metrics to compare
error_metrics = ['MSE', 'RMSE', 'MAE']

# Set the figure size
plt.figure(figsize=(10, 6))

# Create a bar plot for each error metric
for metric in error_metrics:
    plt.bar(df['Model'], df[metric], label=metric)

# Set plot title and labels
plt.title('Error Comparison for Different Regression Models')
plt.xlabel('Model Type')
plt.ylabel('Error Value')

# Add a legend
plt.legend(title='Error Metric')

# Show the plot
plt.savefig('Z:/Research_Projects/Reservoir_Project/ML_outputs/all_error_comparison.png')
plt.show()
```
