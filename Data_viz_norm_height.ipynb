```python
import matplotlib.pyplot as plt
import pandas as pd

# Load data from a CSV file
# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/reservoir_regression_data_normalized.csv') #head of 3000 so only 3000 samples, not the whole dataset 

# Scatter plot for tws against normalized reservoir height
plt.figure(figsize=(10, 6))
plt.scatter(df['normalized_reservoir_height'], df['tws'], color='blue')
plt.xlabel('Normalized Reservoir Height')
plt.ylabel('TWS')
plt.title('TWS vs Normalized Reservoir Height')
plt.grid(True)
plt.show()

# Scatter plot for precipitation against normalized reservoir height
plt.figure(figsize=(10, 6))
plt.scatter(df['normalized_reservoir_height'], df['precipitation'], color='green')
plt.xlabel('Normalized Reservoir Height')
plt.ylabel('Precipitation')
plt.title('Precipitation vs Normalized Reservoir Height')
plt.grid(True)
plt.show()

# Scatter plot for temp against normalized reservoir height
plt.figure(figsize=(10, 6))
plt.scatter(df['normalized_reservoir_height'], df['temp'], color='red')
plt.xlabel('Normalized Reservoir Height')
plt.ylabel('Temperature')
plt.title('Temperature vs Normalized Reservoir Height')
plt.grid(True)
plt.show()
```

```python
#How does different ELEVATION affect the correlation between different features? 
import matplotlib.pyplot as plt
import pandas as pd
import os

# Load data from a CSV file
# Replace 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('Z:/Research_Projects/Reservoir_Project/ML_outputs/reservoir_regression_data_normalized.csv')
df.dropna(inplace=True)

output_directory = 'Z:/Research_Projects/Reservoir_Project/ML_outputs/feature_figures'
os.makedirs(output_directory, exist_ok=True)

unique_elevations = sorted(df['elevation'].unique())

# Get colors based on elevation count
colors = plt.cm.viridis_r([i / len(unique_elevations) for i in range(len(unique_elevations))])

# Scatter plot for tws against normalized reservoir height
plt.figure(figsize=(10, 6))
for i, elevation in enumerate(unique_elevations):
    subset_df = df[df['elevation'] == elevation]
    plt.scatter(subset_df['normalized_reservoir_height'], subset_df['tws'], color=colors[i], label=f'Elevation {elevation}')

plt.xlabel('Normalized Reservoir Height')
plt.ylabel('TWS')
plt.title('TWS vs Normalized Reservoir Height')
plt.legend(title='Elevation', labels=[f'Elevation {elev}' for elev in unique_elevations], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_directory, 'elevation_and_tws_vs_normalized_reservoir_height.png'))
plt.close()

# Scatter plot for precipitation against normalized reservoir height
plt.figure(figsize=(10, 6))
for i, elevation in enumerate(unique_elevations):
    subset_df = df[df['elevation'] == elevation]
    plt.scatter(subset_df['normalized_reservoir_height'], subset_df['precipitation'], color=colors[i], label=f'Elevation {elevation}')

plt.xlabel('Normalized Reservoir Height')
plt.ylabel('Precipitation')
plt.title('Precipitation vs Normalized Reservoir Height')
plt.legend(title='Elevation', labels=[f'Elevation {elev}' for elev in unique_elevations], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_directory, 'elevation_and_precip_vs_normalized_reservoir_height.png'))
plt.close()

# Scatter plot for temp against normalized reservoir height
plt.figure(figsize=(10, 6))
for i, elevation in enumerate(unique_elevations):
    subset_df = df[df['elevation'] == elevation]
    plt.scatter(subset_df['normalized_reservoir_height'], subset_df['temp'], color=colors[i], label=f'Elevation {elevation}')

plt.xlabel('Normalized Reservoir Height')
plt.ylabel('Temperature')
plt.title('Temperature vs Normalized Reservoir Height')
plt.legend(title='Elevation', labels=[f'Elevation {elev}' for elev in unique_elevations], bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(output_directory, 'elevation_and_temp_vs_normalized_reservoir_height.png'))
plt.close()

```
